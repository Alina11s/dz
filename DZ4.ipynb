```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
df = pd.read_csv('kc-house-data.csv')
df = df.drop(['id', 'date', 'sqft_lot', 'zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15'], axis=1)
df.info()
df.head(5)
```

**3.Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости**

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(20,10))
sns.set(font_scale=1.4)
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
plt.title('corr_matrix');
```

**3 Выводы:** 
<br>Больше всего на стоимость влияют размер жилой площади и оценка - чем они больше, тем стоимость недвижимости больше
<br>Также положительно на стоимость влияют количество ванных комнат и вид 
<br>Наименьшая зависимость присутствует между ценой и количеством спален/площадью подвала
<br>Год постройки и наличие реновации никак не связаны со стоимостью недвижимости
